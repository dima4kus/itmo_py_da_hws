### **Задача 1**
**2.8.** В ящике 10 красных и 6 синих пуговиц. Какова вероятность
того, что две наудачу вынутые пуговицы будут одноцветными?

#### **Аналитическое решение**




Всего исходов = 16 * 15 / 2 = 120

Благоприятных исходов:

- Две синих = 6 * 5 / 2 = 15
- Две красных = 10 * 9 / 2 = 45

Тогда, в-сть события C - две наудачу вынутые пуговицы будут одноцветными будет равна

(15 + 45) / 120 = 0.5

#### **Решение с помощью Монте-Карло**

Проведем 10 000 симуляций

```python
red = 10
blue = 6
num_simulations = 100000

same_color_count = 0

box = np.array([0] * red + [1] * blue)

for _ in range(num_simulations):
    chosen_buttons_indices = np.random.choice(len(box), 2, replace=False)
    chosen_buttons = box[chosen_buttons_indices]

    if chosen_buttons[0] == chosen_buttons[1]:
        same_color_count += 1

estimated_probability = same_color_count / num_simulations

print(f"В-сть, что две наудачу вынутые пуговицы будут одноцветными будет равна: {estimated_probability:.4f}")
```

### **Задача 2**

**3.3.** На плоскости начерчены параллельные прямые, находящиеся
друг от друга на расстоянии 2a. На плоскость наудачу брошена игла
длины 2r < 2a. Какова вероятность того, что игла пересечет одну из
прямых?

#### **Аналитическое решение**




В-сть, что игла пересечет одну из
прямых делится на 2 части:

- Знаменатель: все пространство исходов. Это `а * Пи`
- Числитель: площадь благопрятных исходов: Это `2r`

`P = 2r / а * Пи`

#### **Решение с помощью Монте-Карло**

```python
a = 5.0
r = 2.0

num_simulations = 1000000

intersections_count = 0

for _ in range(num_simulations):
    center_to_nearest_line_dist = np.random.uniform(0, a)

    angle = np.random.uniform(0, np.pi)

    reach = r * np.sin(angle)

    if center_to_nearest_line_dist <= reach:
        intersections_count += 1

estimated_probability = intersections_count / num_simulations

print(f"В-сть по Монте-Карло): {estimated_probability:.5f}")

analytical_probability = (2 * r) / (a * np.pi)
print(f"Аналитический способ: {analytical_probability:.5f}")
```
