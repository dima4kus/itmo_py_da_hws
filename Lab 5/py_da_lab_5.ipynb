```python
import pandas as pd
from sklearn.model_selection import train_test_split, KFold
```

```python
df = pd.read_csv('beverage_r.csv', sep=';')
```

```python
df.head()
```

### **Задача 1**
Разделить таблицу данных так, чтобы в обучающую выборку попало 66% наблюдений, а в тестовую выборку 33% наблюдений

```python
train_df, test_df = train_test_split(df, train_size=0.66, random_state=42)
```

```python
print(f"Размер обучающей выборки: {len(train_df)} наблюдений ({len(train_df)/len(df):.2%})")
print(f"Размер тестовой выборки: {len(test_df)} наблюдений ({len(test_df)/len(df):.2%})")
```

### **Задача 2**
Разделить таблицу данных на четыре одинаковых части.

```python
kf = KFold(n_splits=4, shuffle=True, random_state=42)
```

```python
four_parts = []
for i, (train_index, test_index) in enumerate(kf.split(df)):
    part = df.iloc[test_index]
    four_parts.append(part)
```

### **Задача 3**
Разделить таблицу данных на две стратифицированные подвыборки.
По переменным кока-кола и минералка.

```python
df
```

```python
df['strat_group'] = df['COKE'].astype(str) + '_' + df['SPRITE'].astype(str)
```

```python
subsample_1, subsample_2 = train_test_split(
    df, test_size=0.5, stratify=df['strat_group'], random_state=42
)
```

Смотрим сколько наблюдений в каждой подвыборке

```python
print(f"Размер подвыборки 1: {len(subsample_1)} наблюдений")
print(f"Размер подвыборки 2: {len(subsample_2)} наблюдений")
```

Смотрим сколько наблюдений в изначальном дата сете

```python
df['strat_group'].value_counts(normalize=True)
```

Смотрим сколько наблюдений в новых подвыборках

```python
print(subsample_1['strat_group'].value_counts(normalize=True))
print(subsample_2['strat_group'].value_counts(normalize=True))
```
